```python
import psycopg2
import pandas as pd
```

```python
connection = psycopg2.connect(
        host="db",
        port=5432,
        database="patents_db",
        user="postgres",
        password="postgres"
    )
def execute_query(query):
        with connection.cursor() as cursor:
            cursor.execute(query)
            return cursor.fetchall()
def get_claims(patent_id):
    """RETURN ALL CLAIMS FOR A GIVEN PATENT ID ordered by claim number"""
    query = f"SELECT claim_text FROM claims WHERE patent_id = '{patent_id}' ORDER BY claim_number"
    return execute_query(query)
def get_claims_str(patent_id):
    """RETURN ALL CLAIMS FOR A GIVEN PATENT ID ordered by claim number as a single string"""

    claims = get_claims(patent_id)
    return "\n".join([claim[0] for claim in claims])
```

```python
query = "select claim_text from claims order by patent_id, claim_number limit 100"
```

```python
print(execute_query(query))
```

```python
ids = execute_query("select distinct(patent_id) from claims limit 100000")
```

```python
ids = [id[0] for id in ids]
print(ids);
```

```python
length = []
for id in ids:
    l = len(get_claims_str(id).split())
    length.append(l)
    
```

```python
avg_words = sum(length) / len(length)
print(avg_words)
print(max(length))
```
