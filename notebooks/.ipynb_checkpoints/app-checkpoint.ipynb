```python
import psycopg2
import pandas as pd
```

```python
connection = psycopg2.connect(
        host="db",
        port=5432,
        database="patents_db",
        user="postgres",
        password="postgres"
    )
def execute_query(query):
        with connection.cursor() as cursor:
            cursor.execute(query)
            return cursor.fetchall()

```

## App by year (linked with OA)

```python
q = "select year, count(distinct oa.app_id) from office_actions oa join applications app on app.app_id=oa.app_id group by app.year;"
app_by_year = execute_query(q)
print(app_by_year)
```

```python
total = sum(count for _, count in app_by_year)

for year, count in app_by_year:
    percentage = (count / total) * 100
    print(f"{year}: {percentage:.2f}%")
```

## From Valid Citations

```python
q = "select year, count(distinct oa.app_id) from valid_citations oa join applications app on app.app_id=oa.app_id group by app.year;"
app_by_year = execute_query(q)

total = sum(count for _, count in app_by_year)

for year, count in app_by_year:
    percentage = (count / total) * 100
    print(f"{year}: {percentage:.2f}%")
```

## office action date

```python
q = "SELECT EXTRACT(YEAR FROM mail_dt) AS year, COUNT(*) AS count FROM office_actions GROUP BY YEAR ORDER BY YEAR;"

app_by_year = execute_query(q)

total = sum(count for _, count in app_by_year)

for year, count in app_by_year:
    percentage = (count / total) * 100
    print(f"{year}: {percentage:.2f}%")
```

## VIEWS

```python
q = "select count(distinct oa.app_id) from office_actions oa join train_app app on app.app_id=oa.app_id;"
train = execute_query(q)
print('train\t\t', train[0][0])
q = "select count(distinct oa.app_id) from office_actions oa join val_app app on app.app_id=oa.app_id;"
val = execute_query(q)
print('validation\t', val[0][0])
q = "select count(distinct oa.app_id) from office_actions oa join test_app app on app.app_id=oa.app_id;"
test = execute_query(q)
print('test\t\t', test[0][0])

sum = test[0][0]+ train[0][0]+ val[0][0]
print("train:\t",train[0][0]/sum)
print("val:\t",val[0][0]/sum)
print("test:\t",test[0][0]/sum)
```
